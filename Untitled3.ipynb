```python
file_path = "tracks.csv"
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "mcfurland/10-m-beatport-tracks-spotify-audio-features",
    "audio_features.csv"
)

print(df.head())
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['danceability'], df['energy'])
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("Scatter Plot: Danceability vs Energy")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.boxplot(df['loudness'])
plt.xlabel("Loudness")
plt.title("Boxplot of Loudness")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['tempo'])
plt.xlabel("Tempo")
plt.title("Boxplot of Tempo")
plt.show()
```

```python
plt.figure()
plt.boxplot([
    df['danceability'],
    df['energy'],
    df['valence']
])

plt.xticks([1, 2, 3], ['Danceability', 'Energy', 'Valence'])
plt.xlabel("Audio Features")
plt.title("Boxplot of Selected Spotify Audio Features")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['duration_ms'])
plt.xlabel("Duration (ms)")
plt.title("Boxplot of Track Duration")
plt.show()
```
